This clearly indicates a strong and consistent trend of tasks being completed earlier than their due dates. The vertical distance from the line shows the extent of this early completion.*

**On-Time and Minor Delays**: *A smaller proportion of tasks are found either directly on the diagonal line (completed on time) or slightly above it (minor delays).*

**Overall Efficiency**: *The overall pattern suggests a project or team that is highly efficient in its task execution, frequently delivering ahead of schedule. This could imply conservative initial time estimates or excellent workflow management.*

### Feature Engineering

```python
df['expected_days'] = df['due_date'] - df['created_at']
df.expected_days.head()
```

```python
df['expected_days'] = df['expected_days'].dt.days
df['expected_days'].value_counts().plot(kind='bar')

plt.xlabel("Expected Days")
plt.show()
```

#### Given the high occurence of tasks estimated to be completed by 9, 11, 8, 10 days, Lets check how they correlate with the priority of the tasks.

```python
plt.figure(figsize=(12, 5))
plt.subplot(121)
ax = df.groupby('priority')['expected_days'].agg('mean').plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.subplot(122)
ax = df.groupby('category')['expected_days'].agg('mean').plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.show()
```

#### Days vs task priority
- *Critical tasks are expected to be completed within 9.7 days on average, Low priority tasks are eased regarding due dates.*
- *Design categories related tasks takes much more time to complete than the others.*


```python
df['actual_days'] = df['end_time'] - df['start_time']
df['actual_days'] = df['actual_days'].dt.days
df['actual_days'].head()
```

```python
plt.figure(figsize=(14, 5))
plt.subplot(131)
ax = df['actual_days'].value_counts().plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.subplot(132)
ax = df.groupby('priority')['actual_days'].agg('mean').plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.subplot(133)
ax = df.groupby('type')['actual_days'].agg('mean').plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.show()
```

#### Actual days taken to complete tasks
- *Most task completed within 1 day only.*
- *Maximum days taken to complete tasks are 5 days as per available data.*
- *Medium priority tasks takes longer on avergage to complete than others.*
- *There is no definite name specified for the type of tasks that takes longer time to complete.*

```python
plt.figure(figsize=(12, 5))
plt.subplot(121)
ax = df.groupby('category')['actual_days'].agg('mean').plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.title("Actual days taken per category")
plt.subplot(122)
ax = df.groupby('type')['expected_days'].agg('mean').plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.title("Expected days per type")
plt.show()
```

####
- *Research and Development takes around 3 days on average to complete.*
- *Bug and Story types are expected to take more days to complete than average.*
